```python
import matplotlib.pyplot as plt
import cv2
```

```python
image1 = cv2.imread("/media/network/hdd/santosh/cellvit_preprocessing_gall-bladder/patches/S2503154_A8_tiff_pyramid/patches/S2503154_A8_tiff_pyramid_0_34.png")
image2 = cv2.imread("/home/KutumLabGPU/Documents/santosh/capstone_project/models/snake_algo_implement/cellpose/notebook/gall-bladder_rgb_image_with_outlines_0_34.png")
image3 = cv2.imread("/home/KutumLabGPU/Documents/santosh/capstone_project/models/snake_algo_implement/morphsnakes/result/gall-bladder_overlay_0_34.png")
image4 = cv2.imread("/home/KutumLabGPU/Documents/santosh/capstone_project/models/snake_algo_implement/notebook/gall-bladder_patch_with_nuclei_colored_with_legend_0_34.png")

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

# resize
image1 = cv2.resize(image1, (512, 512))
image2 = cv2.resize(image2, (512, 512))
image3 = cv2.resize(image3, (512, 512))
image4 = cv2.resize(image4, (512, 512))

# Create a figure and axes for subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Flatten the axes array for easier indexing
axes = axes.ravel()

# Display images
axes[0].imshow(image1)
axes[0].set_xticks([])  # Hide x-axis ticks
axes[0].set_yticks([])  # Hide y-axis ticks
axes[0].set_title("Original Image", fontsize=14)

axes[1].imshow(image2)
axes[1].set_xticks([])  # Hide x-axis ticks
axes[1].set_yticks([])  # Hide y-axis ticks
axes[1].set_title("Cellpose Segmentation", fontsize=14)

axes[2].imshow(image3)
axes[2].set_xticks([])  # Hide x-axis ticks
axes[2].set_yticks([])  # Hide y-axis ticks
axes[2].set_title("Snake Segmentation", fontsize=14)

axes[3].imshow(image4)
axes[3].set_xticks([])  # Hide x-axis ticks
axes[3].set_yticks([])  # Hide y-axis ticks
axes[3].set_title("Cell ViT Segmentation", fontsize=14)

# Show the plot
plt.suptitle("Image Segmentation Results", fontsize=16)
plt.tight_layout()
plt.show()
```
